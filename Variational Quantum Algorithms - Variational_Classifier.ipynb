```python
#importing PennyLane, the PennyLane-provided version of NumPy, and an optimizer.
import pennylane as qml
from pennylane import numpy as np
from pennylane.optimize import NesterovMomentumOptimizer
```

```python
import matplotlib.pyplot as plt
```

```python
dev = qml.device("default.qubit")   #create a quantum device that will run our circuits
```

```python
###Define One layer of variational circuit
#Variational classifiers usually define a “layer” or “block”,
#which is an elementary circuit architecture that gets repeated to build the full variational circuit.

#will use four qubits, or wires, and consists of an arbitrary rotation on every qubit, as well as a ring of CNOTs that entangles each qubit with its neighbour.

@qml.qnode(dev)
def layer(layer_weights):
    for wire in range(4):
        qml.Rot(*layer_weights[wire], wires=wire)

    for wires in ([0, 1], [1, 2], [2, 3], [3, 0]):
        qml.CNOT(wires)
    return [qml.expval(qml.PauliZ(i)) for i in range(4)]
```

```python
def layer(layer_weights):
    for wire in range(4):
        qml.Rot(*layer_weights[wire], wires=wire)

    for wires in ([0, 1], [1, 2], [2, 3], [3, 0]):
        qml.CNOT(wires)
```

```python
#Define Basis Embedding
#inputs are bitstrings
#The BasisState function provided by PennyLane is made to do just this.

def state_preparation(x):
    qml.BasisState(x, wires=[0, 1, 2, 3])
```

```python
#We want to know what exactly BasisState does
@qml.qnode(dev)
def circuit(x):
    qml.BasisState(x, wires=[0, 1, 2, 3])
    return qml.state()

x = np.array([1, 0, 1, 0])
print(circuit(x))
```

```python
print(qml.draw(circuit, level="device")(x))
```

```python
state = circuit(x)

# Find nonzero index
nonzero_index = np.nonzero(state)[0]

print("Nonzero index:", nonzero_index)
print("Amplitude:", state[nonzero_index])
```

```python
#Variational Circuit
```

```python
#we define the variational quantum circuit as this state preparation routine, followed by a repetition of the layer structure.
@qml.qnode(dev)
def circuit(weights, x):
    state_preparation(x)

    for layer_weights in weights:
        layer(layer_weights)

    return qml.expval(qml.PauliZ(0))
```

```python
#Adding bias as in classical neural networks
def variational_classifier(weights, bias, x):
    return circuit(weights, x) + bias
```

```python
#standard square loss that measures the distance between target labels and model predictions
#Define Loss
def square_loss(labels, predictions):
    # We use a call to qml.math.stack to allow subtracting the arrays directly
    return np.mean((labels - qml.math.stack(predictions)) ** 2)
```

```python
#Define Accuracy
def accuracy(labels, predictions):
    acc = sum(abs(l - p) < 1e-5 for l, p in zip(labels, predictions))
    acc = acc / len(labels)
    return acc
```

```python
def cost(weights, bias, X, Y):
    predictions = [variational_classifier(weights, bias, x) for x in X]
    return square_loss(Y, predictions)
```

```python
#The parity dataset’s train and test sets
data = np.loadtxt("parity_train.txt", dtype=int)
X = np.array(data[:, :-1])
Y = np.array(data[:, -1])
Y = Y * 2 - 1  # shift label from {0, 1} to {-1, 1}

for x,y in zip(X, Y):
    print(f"x = {x}, y = {y}")
```

```python
#We initialize the variables randomly (but fix a seed for reproducibility).
#Remember that one of the variables is used as a bias, while the rest is fed into the gates of the variational circuit.
np.random.seed(0)
num_qubits = 4
num_layers = 2
weights_init = 0.01 * np.random.randn(num_layers, num_qubits, 3, requires_grad=True)
bias_init = np.array(0.0, requires_grad=True)

print("Weights:", weights_init)
print("Bias: ", bias_init)
```

```python
#Next we create an optimizer instance and choose a batch size…
opt = NesterovMomentumOptimizer(0.5)
batch_size = 5
```

```python
#we compute the outputs of the variational classifier and turn them into predictions in {-1,1} by taking the sign of the output.
weights = weights_init
bias = bias_init
for it in range(100):

    # Update the weights by one optimizer step, using only a limited batch of data
    batch_index = np.random.randint(0, len(X), (batch_size,))
    X_batch = X[batch_index]
    Y_batch = Y[batch_index]
    weights, bias = opt.step(cost, weights, bias, X=X_batch, Y=Y_batch)

    # Compute accuracy
    predictions = [np.sign(variational_classifier(weights, bias, x)) for x in X]

    current_cost = cost(weights, bias, X, Y)
    acc = accuracy(Y, predictions)

    print(f"Iter: {it+1:4d} | Cost: {current_cost:0.7f} | Accuracy: {acc:0.7f}")
```

```python
data = np.loadtxt("parity_test.txt", dtype=int)
X_test = np.array(data[:, :-1])
Y_test = np.array(data[:, -1])
Y_test = Y_test * 2 - 1  # shift label from {0, 1} to {-1, 1}

predictions_test = [np.sign(variational_classifier(weights, bias, x)) for x in X_test]

for x,y,p in zip(X_test, Y_test, predictions_test):
    print(f"x = {x}, y = {y}, pred={p}")

acc_test = accuracy(Y_test, predictions_test)
print("Accuracy on unseen data:", acc_test)
```
